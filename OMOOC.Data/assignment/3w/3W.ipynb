**1. 二维随机漫步**

```python
%matplotlib inline
import numpy as np
import random
import matplotlib as mpl
import matplotlib.pyplot as plt

position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)

fig, axes = plt.subplots(1,4,figsize=(16,4))
for n in range(4):
    axes[n].plot(walk,color="r",label=r"$Walk trace$")
    
    axes[n].legend(loc=0,fontsize=12)    
    
    axes[n].set_xticks(np.linspace(0,1000,6))
    axes[n].set_yticks(np.linspace(-50,50,11))
    
    axes[n].set_xlabel("Steps",fontsize=15)
    axes[n].set_ylabel("Trace",fontsize=15)
    
    axes[n].set_title("Trace of a Drunk man (%d)" %(n+1),fontsize=15);

fig.tight_layout()
```

**2. 二次函数，梯形法求定积分的各个梯形**

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return 3 -2*x - x**2

fig,ax = plt.subplots(figsize=(8,4))

def plot_axes(x,a,b,n,f):
    x = np.linspace(a,b,100)
    ax.plot(x,f(x), color="black", lw=1)
    ax.plot(x,np.zeros_like(x), lw=1, color="black")
    ax.set_xticks(np.linspace(a,b,(b-a+1)))
    
    for i in range(n):
        x = np.linspace(a,b,n)
        ax.plot([x[i],x[i]],[0,f(x[i])], lw=1.5, ls="--", color="r", alpha=0.5)
        ax.scatter(x[i],f(x[i]),10,color="r")
        if i==0: continue
        ax.plot([x[i-1],x[i]],[f(x[i-1]),f(x[i])], color="r", ls="--", lw=2.5, alpha=0.5)

plot_axes(x,-5,5,10,f)
```

**3.交互式函数**

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed

def f(n,m):
    x = np.linspace(-10,10,m)
    y = np.sin(x)
    fig,axes = plt.subplots(1,2,figsize=(8,4),sharey=True)
    axes[0].hist(y,bins=n)
    axes[1].plot(x,y)

interact(f,n=(1,10),m=(1,50)); #n控制直方图条数，m控制自变量的密集程度
```
