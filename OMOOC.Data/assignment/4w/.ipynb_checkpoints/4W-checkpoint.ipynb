```python
from scipy.optimize import leastsq
import pylab as pl

def residuals(p):
    k, b = p
    return y - (k*x + b)

r = leastsq(residuals,[1,0])
k, b = r[0]
print "Volume =", k,"*Open + ", b

xx = np.linspace(2500,6000,1000)
yy = xx*k + b

fig, ax = plt.subplots(1,1,figsize=(8,4))

ax.scatter(x,y)
ax.plot(xx,yy)

ax.set_xlabel("Open",fontsize=12)
ax.set_ylabel("Volume (Million)",fontsize=12);
```

###  Step3: 画分布图

```python
eq2.head()
```

```python
lons, lats = list(eq2['lon']), list(eq2['lat'])
mags = list(eq2['m_num'])
```

```python
lons, lats = list(eq2['lon']), list(eq2['lat'])
mags = list(eq2['m_num'])

def get_marker_color(magnitude):
    if magnitude < 3.0:
        return ('go')
    elif magnitude < 5.0:
        return ('yo')
    else:
        return ('ro')

fig, ax = plt.subplots(1,1,figsize=(20,15))
    
eq_map = Basemap(projection='robin', resolution='l',area_thresh = 1000.0,
                lat_0=0, lon_0=120)
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color = 'gray')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0,360,30))
eq_map.drawparallels(np.arange(-90,90,30))

min_marker_size = 2.5

for lon, lat, mag in zip(lons, lats, mags):
    x,y = eq_map(lon,lat)
    msize = mag * min_marker_size
    marker_string = get_marker_color(mag)
    eq_map.plot(x, y, marker_string, markersize=msize)
    
plt.show()
```

```python
#发现国内数据较多，我们放大地图，缩小范围。

llcrnrlon, llcrnrlat = eq2['lon'].min(), eq2['lat'].min()
urcrnrlon, urcrnrlat = eq2['lon'].max(), eq2['lat'].max()

lons, lats = list(eq2['lon']), list(eq2['lat'])
mags = list(eq2['m_num'])

def get_marker_color(magnitude):
    if magnitude < 3.0:
        return ('go')
    elif magnitude < 5.0:
        return ('yo')
    else:
        return ('ro')

fig, ax = plt.subplots(1,1,figsize=(20,15))
    
eq_map = Basemap(projection='merc', resolution = 'l', area_thresh = 1000.0,
                lat_0=0, lon_0=120,
                llcrnrlon=70, llcrnrlat=0, urcrnrlon=140, urcrnrlat=55) #ortho 球  robin 全球平面图  merc 部分图
eq_map.drawcoastlines()
eq_map.drawcountries()
eq_map.fillcontinents(color = 'lightgray')
eq_map.drawmapboundary()
eq_map.drawmeridians(np.arange(0,360,30))
eq_map.drawparallels(np.arange(-90,90,30))

min_marker_size = 4

for lon, lat, mag in zip(lons, lats, mags):
    x,y = eq_map(lon,lat)
    msize = mag * min_marker_size
    marker_string = get_marker_color(mag)
    eq_map.plot(x, y, marker_string, markersize=msize)
    
plt.show()
```
