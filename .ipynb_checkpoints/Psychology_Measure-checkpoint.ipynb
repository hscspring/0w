## 信度

主要针对信度（内部一致性）：  

- alpha 系数
- 折半信度

主要分四个方面：

- 整体所有类别
- 整体不同类别
- 不同年龄段所有类别
- 不同年龄段不同类别

```python
# 不同年龄段不同类别
# 不需要只按类别，只按类别，没有用户满足（意味着不同年龄段用户做的类别不同）
outdir = "/Users/HaoShaochun/Desktop/output/csv/"
cate_ques = dict()
for age_inte in  sorted(list(set(clean_df.age_interval)))[:-1]:
    # 必须分年龄段，
    # 如果用所有年龄段，没有一个类别有用户全部做了
    for ll2 in list(cates_level23.columns):
        # 只能用 level2
        # level1，没有一个用户全部做了
        # level3，只有一道题，无法使用
        cate_df = clean_df[clean_df.age_interval==age_inte]
        cate_df = cate_df[cate_df.level2==ll2]
        
        # 剔除没做所有题目的 用户
        scores = cate_df.pivot(index="id", columns="question", values='answer')
        scores = scores.drop([idindex for idindex in scores.index 
                                  if None in list(scores.T[idindex])])
        
        # z标准化数据
        score_array = np.array(scores, dtype=np.int)
        mu = np.average(score_array, axis=0)
        sigma = np.std(score_array, axis=0)
        std_zscores = (scores - mu)/(sigma+0.1)
        
        # 0-1标准化数据
        maxx = score_array.max(axis=0)
        minn = score_array.min(axis=0)
        std_01scores = (scores - minn)/maxx
        
        ###############计算指标###############
        
        # 原始数据计算
        alpha = BaseCtt(scores).get_alpha_reliability()
        halfscore = get_half_reliability(scores)
        # 标准化数据计算
        stdz_alpha = BaseCtt(std_zscores).get_alpha_reliability()
        stdz_halfscore = get_half_reliability(std_zscores)
        # 标准化数据计算2
        std01_alpha =  BaseCtt(std_01scores).get_alpha_reliability()
        std01_halfscore =  get_half_reliability(std_01scores)
        
        ###############数据存储###############
        cate_ques[age_inte + ll2] = list(scores.columns)
        out_scores = scores
        out_scores.columns = ["Q"+str(i+1) for i in range(len(out_scores.columns))]
#         out_scores.columns = ["".join(pnormal.findall(text)) for text in 
#                               list(out_scores.columns)]
        out_scores.to_csv(outdir + age_inte + ll2 + ".csv")
        
        
        print(age_inte, ll2)
        print("用户数 * 该类别题数：", scores.shape)
        print("alpha 系数: 原始得分 %.2f, z得分 %.2f, 01得分 %.2f" % (alpha, 
                                                             stdz_alpha,
                                                            std01_alpha))
        print("折半信度系数: 原始得分 %.2f, z得分 %.2f, 01得分 %.2f" % (halfscore, 
                                               stdz_halfscore,
                                              std01_halfscore))
        print()
```
